```python
import pandas as pd

df = pd.read_csv("silver_prices_historical.csv",parse_dates=["Date"],index_col="Date")
```

```python
### Visualization
df.head()
```

```python
cols = df.columns
for col in cols:
    print(f"type({col} = {type(col)})")
```

```python
import matplotlib.pyplot as plt
monthly_close_mean = df["Close"].resample("ME").mean()
monthly_open_mean = df["Open"].resample("ME").mean()

plt.plot(monthly_close_mean,label="Monthly Close Mean")
plt.plot(monthly_open_mean,label="Monthly Open Mean")
plt.legend()
plt.show()
```
