```python
import requests as re
from bs4 import BeautifulSoup

url = "https://quotes.toscrape.com/tag/humor"
response = re.get(url)
html = response.text

soup = BeautifulSoup(html,"html.parser")
```

```python
H1 = soup.find_all("h1")
H2 = soup.find_all("h2")
H3 = soup.find_all("h3")
P = soup.find_all("p")
SPAN = soup.find_all("span",class_="text")
SMALL = soup.find_all("small")
```

```python
for h1 in H1:
    print(h1.get_text())
```

```python
for h2 in H2:
    print(h2.get_text())
```

```python
for p in P:
    print(p.get_text())
```

```python
for span in SPAN:
    print(span.get_text())
```

```python
for small in SMALL:
    print(small.get_text())
```

```python
quotes, author = [], []
```

```python
for i in range(len(SMALL)):
    print("Quote: ",SPAN[i].get_text())
    quotes.append(SPAN[i].get_text())
    print("Author: ",SMALL[i].get_text())
    author.append(SMALL[i].get_text())
    print("\n")
```

```python
import pandas as pd
quotes_dict = {
    "author": author,
    "quotes": quotes

}
df = pd.DataFrame(quotes_dict)
df.to_csv("quotes.csv",index=False)
```
