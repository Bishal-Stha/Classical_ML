```python
import pandas as pd

df = pd.read_csv("silver_prices_forecast_2026.csv",parse_dates=["Date"],index_col="Date")
# parse_dates: because without it, Date would a string wouldn't work for indexing.
# index_col="Date" because now, Date is the index and we can easily slice on that.
df.head()
```

```python
df.loc["2026-01"].Lower_Bound.mean()
```

```python
df.loc["2026-01-01":"2026-01-21"].describe()
```

```python
## Resampling
df.resample("W").mean() # resampling divides the data into timePeriod. based on Symbols. like W is week, ME is month, etc.
```

```python
%matplotlib inline
df["Upper_Bound"].resample("M").mean().plot(kind="bar")
```
