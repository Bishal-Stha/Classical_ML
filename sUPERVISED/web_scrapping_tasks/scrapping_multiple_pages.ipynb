```python
QUOTES, AUTHORS = [], []
```

```python
for i in range(10):
    URL = url+str(i+1)
    response = re.get(URL)
    if(response.status_code != 200):
        break
    html = response.text
    soup = BeautifulSoup(html,"html.parser")
    quotes = soup.find_all("span",class_="text")
    authors = soup.find_all("small",class_="author")

    # [QUOTES.append(quote.get_text()) for quote in quotes]
    for quote in quotes:
        QUOTES.append(quote.get_text())
    # [AUTHORS.append(author.get_text()) for author in authors]
    for author in authors:
        AUTHORS.append(author.get_text())
```

```python
import pandas as pd

quotes_dict = {
    "author": AUTHORS,
    "quote": QUOTES
}

df = pd.DataFrame(quotes_dict)
df.to_csv("all_quotes.csv",index=False)
```
