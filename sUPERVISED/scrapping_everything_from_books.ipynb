```python
import pandas as pd
import requests
# import re
from bs4 import BeautifulSoup
import time
from urllib.parse import urljoin

BASE_URL = "https://books.toscrape.com/"
current_url = BASE_URL
session = requests.Session()
```

```python
rate = {
    "One": 1,
    "Two": 2,
    "Three": 3,
    "Four": 4,
    "Five": 5
}

TITLE = []
RATING = []
CATEGORY = []
PRICE = []
DESCRIPTION = []
AVAILABILITY = []
```

```python
while True:
    response = session.get(current_url)
    if response.status_code != 200:
        break

    soup = BeautifulSoup(response.text, "html.parser")
    books = soup.find_all("article",class_="product_pod")
    for book in books:
        title = book.find("h3").find("a")["title"]
        price = book.find("p", class_="price_color").get_text(strip=True)
        rating = book.find("p",class_="star-rating")["class"][1]
        available = book.find("p",class_="instock availability").get_text().strip()
        available = " ".join(available.split())

        book_url = book.find("a")["href"]
        book_desc_url = urljoin(BASE_URL, book_url)

        nested_response = session.get(book_desc_url)
        doubleSoup = BeautifulSoup(nested_response.text, "html.parser")

        breadcrumbs = doubleSoup.select("ul.breadcrumb li a")
        category = breadcrumbs[2].get_text(strip=True) if len(breadcrumbs) > 2 else "Unknown"
        
        desc_tag = doubleSoup.select_one("#product_description ~ p")
        desc = desc_tag.get_text(strip=True) if desc_tag else "No description"


        # Appending
        TITLE.append(title)
        PRICE.append(price)
        RATING.append(rating)
        AVAILABILITY.append(available)
        CATEGORY.append(category)
        DESCRIPTION.append(desc)

    next_btn = soup.find("li", class_="next")
    if next_btn is None:
        print("Reached the last page.")
        break

    next_page = next_btn.find("a")["href"]
    current_url = urljoin(current_url, next_page)


    time.sleep(1)




```

```python
df = pd.DataFrame({
    "title": TITLE,
    "price": PRICE,
    "category": CATEGORY,
    "description": DESCRIPTION,
    "avalability": AVAILABILITY,
    "rating": RATING
})

df["price"] = df["price"].str.replace("Ã‚", "", regex= False)
df.to_csv("Scrapped_books.csv")
```
