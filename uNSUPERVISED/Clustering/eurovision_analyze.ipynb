```python
df["Jury Points"] = df["Jury Points"].fillna(df["Jury Points"].mean())
df["Televote Points"] = df["Televote Points"].fillna(df["Televote Points"].mean())
df.isna().sum()
```

```python
pca = PCA(n_components=2)
scaler = StandardScaler()
df_scaled = scaler.fit_transform(df)
df_pca_scaled = pca.fit_transform(df_scaled)
df_pca_scaled.shape
```
