```python
import pandas as pd
from sklearn.cluster import DBSCAN, dbscan
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler

df = pd.read_csv("../data/Wholesale customers data.csv")
dbscan = DBSCAN(eps=0.5, min_samples=5, metric='euclidean')
scaler = StandardScaler()
pca = PCA(n_components=2)

df_scaled = scaler.fit_transform(df)
df_pca_scaled = pca.fit_transform(df_scaled)
print(f"Scaled df shape: {df_scaled.shape}")
print(f"Scaled pca df shape: {df_pca_scaled.shape}")

dbscan.fit(df_pca_scaled)
```

```python
dbscan.labels_
```

```python
import matplotlib.pyplot as plt
plt.scatter(df_pca_scaled[:,0],df_pca_scaled[:,1],c=dbscan.labels_)
```

```python
from sklearn.metrics import silhouette_score
score = silhouette_score(df_pca_scaled,dbscan.labels_)
print(f"Silhouette Score: {score:.3f}")
```
